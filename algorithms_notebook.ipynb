# Algorithms

```python
def fibonacci_non_recurcive(n : int):
    if n <= 1:
        return n
    prev, cur = 0, 1
    for _ in range(n-1):
        prev, cur = cur, prev + cur
    return cur

[fibonacci_non_recurcive(i) for i in range(15)]
```

```python
def numbers(n):
    # print(n)
    if n <= 0:
        print('Starting returning...')
        return
    else:
        print('Forward to stack = ', n)
        numbers(n-1)
        print('Back from stack', n)

numbers(3)
```

```python
def factorial(n :int):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return factorial(n-1) * n
    
[factorial(n) for n in range(10)]
```

Ниже показан рекурсивный алгоритм генерации всех скобочных последовательностей с использованием традиционного стека, где элементы вкладываются и достаются из стека до и после вызова функции внутри функции.

```python
def generate(n):
    result = []

    def generate_(left_open, left_closed, accum):
        if not left_open and not left_closed:
            result.append(''.join(accum))
            return
        if left_open:
            accum.append('(')
            generate_(left_open - 1, left_closed, accum)
            accum.pop()
        if left_closed > left_open:
            accum.append(')')
            generate_(left_open, left_closed - 1, accum)
            accum.pop()

    generate_(n, n, [])
    return result

generate(3)
```

```python
def rec_arr_sum(l:list[float]):
    if l == []:
        return 0
    return l[0] + rec_arr_sum(l[1:])

rec_arr_sum([1.5,2,3,4,3,2,1])
```

```python
def rec_elem_count(l: list):
    if l == []:
        return 0
    return 1 + rec_elem_count(l[1:])

rec_elem_count([1,4,3,5,2,1])
```

```python
def rec_max(l: list):
    if len(l) == 2:
        return l[0] if l[0] > l[1] else l[1]
    sub_max = rec_max(l[1:])
    return l[0] if l[0] > sub_max else sub_max

rec_max([1,2,6,34,86,32,1,2,5,2,76,34,22])
```

```python
import random

def q_sort(l:list[float]):
    if len(l) < 2:
        return l
    pivot_index = random.randint(0, len(l) - 1)
    pivot = l[pivot_index]
    less = [x for x in l if x <= pivot and x != l[pivot_index]]
    greater = [x for x in l if x > pivot]
    return q_sort(less) + [pivot] + q_sort(greater)


q_sort([1,23,53,2,43,357,8765,343,4,234,42,54,2,4,545,23456,65654])
```
